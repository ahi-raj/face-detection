```python
import cv2
```

```python
faceCascade=cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
image=cv2.imread("img.jpg")
```

```python
from matplotlib import pyplot as plt
plt.imshow(image)
```

```python
RGB_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
GR_img=cv2.cvtColor(RGB_img,cv2.COLOR_BGR2GRAY)
plt.imshow(GR_img)
```

```python
faces=faceCascade.detectMultiScale(
    GR_img,
    scaleFactor=1.5,
    minNeighbors=5,
    minSize=(20,20)
)

print(faces)
```

```python
for x, y, w, h in faces:
    image=cv2.rectangle(image,(x,y),(x+w,y+h),(0,0,255),10)
RGB_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
resized=cv2.resize(image,(int(image.shape[1]/3), int(image.shape[0]/3)))
```

```python
cv2.imshow("Deteced-face",resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
